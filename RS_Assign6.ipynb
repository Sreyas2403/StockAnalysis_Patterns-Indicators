```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nsepython as nse
import matplotlib.dates as mdates

symbol = input()
series = "EQ"
start_date = input()
end_date = input()
equity_data = equity_history(symbol, series, start_date, end_date)

# Convert the equity data to a DataFrame
equity_df = pd.DataFrame(equity_data)

# Convert the 'date' column to datetime format
equity_df['mTIMESTAMP'] = pd.to_datetime(equity_df['mTIMESTAMP'])

# Sort the data by date
equity_df = equity_df.sort_values(by='mTIMESTAMP')

# Calculate daily returns
equity_df['price_change'] = equity_df['CH_CLOSING_PRICE'].diff()

# Calculate the gains and losses
equity_df['gain'] = np.where(equity_df['price_change'] > 0, equity_df['price_change'], 0)
equity_df['loss'] = np.where(equity_df['price_change'] < 0, -equity_df['price_change'], 0)

# Calculate the average gain and average loss
rolling_window = 14
equity_df['avg_gain'] = equity_df['gain'].rolling(window=rolling_window, min_periods=1).mean()
equity_df['avg_loss'] = equity_df['loss'].rolling(window=rolling_window, min_periods=1).mean()

# Calculate the Relative Strength (RS)
equity_df['RS'] = equity_df['avg_gain'] / equity_df['avg_loss']

print(equity_df['RS'])
# Print the output RS values into an excel file
output_file = 'SBIN_RS_values.xlsx'
equity_df.to_excel(output_file, columns=['mTIMESTAMP', 'RS'], index=False)

# Plot the RS values against dates in a line graph
plt.figure(figsize=(15, 6))
plt.plot(equity_df['mTIMESTAMP'], equity_df['RS'], label='Relative Strength (RS)', color='bLUE')
plt.xlabel('Date')
plt.ylabel('Relative Strength (RS)')
plt.title(f'Relative Strength (RS) of {symbol}')
plt.legend()
plt.grid(True)

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=20))

# Rotate date labels for better readability
plt.gcf().autofmt_xdate()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nsepython as nse
import matplotlib.dates as mdates

# Inputs
symbol = input()
series = "EQ"
start_date = input()
end_date = input()

# Fetching equity data
equity_data_sbin = nse.equity_history(symbol, series, start_date, end_date)
equity_data_nifty = nse.equity_history("NIFTY", series, start_date, end_date)

# Convert data to DataFrame
data_sbin = pd.DataFrame(equity_data_sbin)
data_nifty = pd.DataFrame(equity_data_nifty)

# Ensure both dataframes are sorted by date and indexed by date for accurate RS calculation
data_sbin['mTIMESTAMP'] = pd.to_datetime(data_sbin['mTIMESTAMP'])
data_nifty['mTIMESTAMP'] = pd.to_datetime(data_nifty['mTIMESTAMP'])
data_sbin.set_index('mTIMESTAMP', inplace=True)
data_nifty.set_index('mtimstamp', inplace=True)

# Calculate RS (Relative Strength)
data_sbin['RS'] = data_sbin['CH_CLOSING_PRICE'] / data_nifty['CH_CLOSING_PRICE']

# Print RS values into an Excel file
output_file = 'SBIN_Relative_Strength.xlsx'
data_sbin.to_excel(output_file, columns=['RS'])

# Plot RS values against dates
plt.figure(figsize=(14, 7))
plt.plot(data_sbin.index, data_sbin['RS'], label='Relative Strength (SBIN/NIFTY50)', color='blue')
plt.xlabel('Date')
plt.ylabel('Relative Strength')
plt.title('Relative Strength (RS) of SBIN relative to NIFTY50')
plt.legend()
plt.grid(True)

# Format x-axis to show more dates
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.gcf().autofmt_xdate()

# Show plot
plt.show()

print(f"RS values have been saved to {output_file}")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Define the stock and benchmark symbols
stock_symbol = 'SBIN.NS'  # NSE symbol for State Bank of India in yfinance
benchmark_symbol = '^NSEI'  # NSE symbol for NIFTY 50

# Define the date range
start_date = '2023-01-01'
end_date = '2023-12-31'

# Fetch historical data for the stock and the benchmark index
stock_data = yf.download(stock_symbol, start=start_date, end=end_date)
benchmark_data = yf.download(benchmark_symbol, start=start_date, end=end_date)

# Calculate the relative strength
relative_strength = stock_data['Close'] / benchmark_data['Close']

# Combine the data into a single DataFrame for better visualization
rs_df = pd.DataFrame({
    'Stock Close': stock_data['Close'],
    'Benchmark Close': benchmark_data['Close'],
    'Relative Strength': relative_strength})

# Display the first few rows of the DataFrame
print(rs_df)

# Save the data to a CSV file (optional)
rs_df.to_csv('relative_strength_SBIN_vs_NIFTY50.csv', index=True)

plt.figure(figsize=(10, 6))
plt.plot(rs_df.index, rs_df['Relative Strength'], label='Relative Strength (SBIN vs NIFTY50)', color='blue')
plt.xlabel('Date')
plt.ylabel('Relative Strength')
plt.title('Relative Strength of SBIN Relative to NIFTY50 (2023)')
plt.legend()
plt.grid(True)


plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=20))

# Rotate date labels for better readability
plt.gcf().autofmt_xdate()
plt.show()
```
