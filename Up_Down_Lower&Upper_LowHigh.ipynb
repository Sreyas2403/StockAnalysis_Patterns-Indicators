```python
import pandas as pd
import nsepython as nse
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Define symbols and series
stock_symbol = "SBIN"
series = "EQ"
start_date = input()
end_date = input()

# Fetch equity history for SBIN
equity_data = equity_history(stock_symbol, series, start_date, end_date)
stock_df = pd.DataFrame(equity_data)
stock_df['mTIMESTAMP'] = pd.to_datetime(stock_df['mTIMESTAMP'])
stock_df = stock_df.sort_values(by='mTIMESTAMP')

# Convert necessary columns to numeric
stock_df['CH_CLOSING_PRICE'] = pd.to_numeric(stock_df['CH_CLOSING_PRICE'], errors='coerce')
stock_df['CH_OPENING_PRICE'] = pd.to_numeric(stock_df['CH_OPENING_PRICE'], errors='coerce')
stock_df['CH_TRADE_HIGH_PRICE'] = pd.to_numeric(stock_df['CH_TRADE_HIGH_PRICE'], errors='coerce')
stock_df['CH_TRADE_LOW_PRICE'] = pd.to_numeric(stock_df['CH_TRADE_LOW_PRICE'], errors='coerce')

# Drop rows with NaN values in key columns after conversion
stock_df = stock_df.dropna(subset=['CH_CLOSING_PRICE', 'CH_OPENING_PRICE', 'CH_TRADE_HIGH_PRICE', 'CH_TRADE_LOW_PRICE'])

# Function to identify uptrends
def identify_uptrends(df):
    uptrends = []
    higher_highs = []
    higher_lows = []
    start_closing_prices = []
    end_closing_prices = []

    start_date = None
    highest_high = None
    lowest_low = None
    start_closing_price = None

    for i in range(1, len(df)):
        if df['CH_CLOSING_PRICE'][i] > df['CH_CLOSING_PRICE'][i - 1]:
            if start_date is None:
                start_date = df['mTIMESTAMP'][i - 1]
                highest_high = df['CH_TRADE_HIGH_PRICE'][i - 1]
                lowest_low = df['CH_TRADE_LOW_PRICE'][i - 1]
                start_closing_price = df['CH_CLOSING_PRICE'][i - 1]
                end_closing_price = df['CH_CLOSING_PRICE'][i]
                highest_high = max(highest_high, df['CH_TRADE_HIGH_PRICE'][i])
                lowest_low = min(lowest_low, df['CH_TRADE_LOW_PRICE'][i])
        else:
            if start_date is not None:
                end_date = df['mTIMESTAMP'][i - 1]
                end_closing_price = df['CH_CLOSING_PRICE'][i - 1]
                uptrends.append((start_date, end_date))
                higher_highs.append(highest_high)
                higher_lows.append(lowest_low)
                start_closing_prices.append(start_closing_price)
                end_closing_prices.append(end_closing_price)
                start_date = None
                highest_high = None
                lowest_low = None

    if start_date is not None:
        uptrends.append((start_date, df['mTIMESTAMP'].iloc[-1]))
        higher_highs.append(highest_high)
        higher_lows.append(lowest_low)
        start_closing_prices.append(start_closing_price)
        end_closing_prices.append(df['CH_CLOSING_PRICE'].iloc[-1])

    return uptrends, higher_highs, higher_lows, start_closing_prices, end_closing_prices

# Calculate uptrends
uptrends, higher_highs, higher_lows, start_closing_prices, end_closing_prices  = identify_uptrends(stock_df)

# Print results
print("Uptrends (Start Date, End Date):")
for start, end in uptrends:
    print(f"({start.date()}, {end.date()})")

print("\nHigher Highs:")
print(higher_highs)

print("\nHigher Lows:")
print(higher_lows)

#saving to an excel
uptrends_data = {
    'Start Date': [start.date() for start, end in uptrends],
    'End Date': [end.date() for start, end in uptrends],
    'Higher High': higher_highs,
    'Higher Low': higher_lows,
    'Start Closing Price': start_closing_prices,
    'End Closing Price': end_closing_prices
}
uptrends_df = pd.DataFrame(uptrends_data)

excel_file_path = 'SBIN_uptrends.xlsx'
uptrends_df.to_excel(excel_file_path, index=False)

# Plotting  the graph of closing price along with the identified uptrends
plt.figure(figsize=(14, 7))
plt.plot(stock_df['mTIMESTAMP'], stock_df['CH_CLOSING_PRICE'], label='Closing Price', color='blue')

for start, end in uptrends:
    plt.axvspan(start, end, color='green', alpha=0.3)


plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title(f'Uptrends in {stock_symbol}')
plt.legend()
plt.grid()

# Formatting date labels
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gcf().autofmt_xdate()

plt.show()
```

```python
import pandas as pd
import nsepython as nse
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Define symbols and series
symbol = "SBIN"
series = "EQ"
start_date = input()
end_date = input()

# Fetch equity history for SBIN
equity_data = nse.equity_history(symbol, series, start_date, end_date)
stock_df = pd.DataFrame(equity_data)
stock_df['mTIMESTAMP'] = pd.to_datetime(stock_df['mTIMESTAMP'])
stock_df = stock_df.sort_values(by='mTIMESTAMP')

# Convert necessary columns to numeric
stock_df['CH_CLOSING_PRICE'] = pd.to_numeric(stock_df['CH_CLOSING_PRICE'], errors='coerce')
stock_df['CH_OPENING_PRICE'] = pd.to_numeric(stock_df['CH_OPENING_PRICE'], errors='coerce')
stock_df['CH_TRADE_HIGH_PRICE'] = pd.to_numeric(stock_df['CH_TRADE_HIGH_PRICE'], errors='coerce')
stock_df['CH_TRADE_LOW_PRICE'] = pd.to_numeric(stock_df['CH_TRADE_LOW_PRICE'], errors='coerce')

# Drop rows with NaN values in key columns after conversion
stock_df = stock_df.dropna(subset=['CH_CLOSING_PRICE', 'CH_OPENING_PRICE', 'CH_TRADE_HIGH_PRICE', 'CH_TRADE_LOW_PRICE'])

# Function to calculate RSI
def calculate_rsi(data, window=14):
    delta = pd.Series(data).diff(1)
    gain = (delta.where(delta > 0, 0)).rolling(window=window, min_periods=1).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window, min_periods=1).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Calculate RSI
stock_df['RSI'] = calculate_rsi(stock_df['CH_CLOSING_PRICE'])

# Function to identify divergences
def identify_divergences(df):
    positive_divergences = []
    negative_divergences = []
    for i in range(2, len(df)):
        # Positive Divergence
        if df['CH_TRADE_LOW_PRICE'][i] < df['CH_TRADE_LOW_PRICE'][i - 1] or df['CH_TRADE_LOW_PRICE'][i - 1] < df['CH_TRADE_LOW_PRICE'][i - 2]:
            if df['RSI'][i] > df['RSI'][i - 1] or df['RSI'][i - 1] > df['RSI'][i - 2]:
                positive_divergences.append((df['mTIMESTAMP'][i], df['CH_TRADE_LOW_PRICE'][i], df['RSI'][i]))
        # Negative Divergence
        if df['CH_TRADE_HIGH_PRICE'][i] > df['CH_TRADE_HIGH_PRICE'][i - 1] or df['CH_TRADE_HIGH_PRICE'][i - 1] > df['CH_TRADE_HIGH_PRICE'][i - 2]:
            if df['RSI'][i] < df['RSI'][i - 1] or df['RSI'][i - 1] < df['RSI'][i - 2]:
                negative_divergences.append((df['mTIMESTAMP'][i], df['CH_TRADE_HIGH_PRICE'][i], df['RSI'][i]))
    return positive_divergences, negative_divergences

# Identify divergences
positive_divergences, negative_divergences = identify_divergences(stock_df)

# Print detected divergences
print("Positive Divergences:", positive_divergences)
print("Negative Divergences:", negative_divergences)

# Prepare DataFrame to save to Excel
positive_divergence_data = {
    'Date': [date for date, price, rsi in positive_divergences],
    'Price Low': [price for date, price, rsi in positive_divergences],
    'RSI High': [rsi for date, price, rsi in positive_divergences]
}
negative_divergence_data = {
    'Date': [date for date, price, rsi in negative_divergences],
    'Price High': [price for date, price, rsi in negative_divergences],
    'RSI Low': [rsi for date, price, rsi in negative_divergences]
}
positive_divergence_df = pd.DataFrame(positive_divergence_data)
negative_divergence_df = pd.DataFrame(negative_divergence_data)

# Save to Excel file
with pd.ExcelWriter('SBIN_RSI_Divergences.xlsx') as writer:
    positive_divergence_df.to_excel(writer, sheet_name='Positive Divergences', index=False)
    negative_divergence_df.to_excel(writer, sheet_name='Negative Divergences', index=False)

# Print confirmation
print("Positive and negative divergence data saved to SBIN_divergences.xlsx")

# Plot the closing price and RSI with divergences highlighted
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 10), sharex=True)

# Plot Closing Price
ax1.plot(stock_df['mTIMESTAMP'], stock_df['CH_CLOSING_PRICE'], label='Closing Price', color='blue')
ax1.set_ylabel('Closing Price')
ax1.set_title(f'Positive Divergences in {symbol}')
ax1.legend()
ax1.grid()

# Highlight positive divergences on price plot
for date, price, rsi in positive_divergences:
    ax1.annotate('Positive Divergence', xy=(date, price), xytext=(date, price + 10),
                 arrowprops=dict(facecolor='green', shrink=0.05))

# Highlight negative divergences on price plot
for date, price, rsi in negative_divergences:
    ax1.annotate('Negative Divergence', xy=(date, price), xytext=(date, price - 10),
                 arrowprops=dict(facecolor='red', shrink=0.05))

# Plot RSI
ax2.plot(stock_df['mTIMESTAMP'], stock_df['RSI'], label='RSI', color='purple')
ax2.axhline(70, linestyle='--', alpha=0.5, color='gray')
ax2.axhline(30, linestyle='--', alpha=0.5, color='gray')
ax2.set_ylabel('RSI')
ax2.set_xlabel('Date')
ax2.legend()
ax2.grid()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gcf().autofmt_xdate()
# Highlight positive divergences on RSI plot
for date, price, rsi in positive_divergences:
    ax2.annotate('Positive Divergence', xy=(date, rsi), xytext=(date, rsi + 5),
                 arrowprops=dict(facecolor='green', shrink=0.05))

# Highlight negative divergences on RSI plot
for date, price, rsi in negative_divergences:
    ax2.annotate('Negative Divergence', xy=(date, rsi), xytext=(date, rsi - 5),
                 arrowprops=dict(facecolor='red', shrink=0.05))

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import nsepython as nse
import matplotlib.pyplot as plt

# Define symbols and series
symbol = input()
series = "EQ"
start_date = input()
end_date = input()

# Fetch equity history for SBIN
equity_data = nse.equity_history(symbol, series, start_date, end_date)
stock_df = pd.DataFrame(equity_data)
stock_df['mTIMESTAMP'] = pd.to_datetime(stock_df['mTIMESTAMP'])
stock_df = stock_df.sort_values(by='mTIMESTAMP')

# Convert necessary columns to numeric
stock_df['CH_CLOSING_PRICE'] = pd.to_numeric(stock_df['CH_CLOSING_PRICE'], errors='coerce')
stock_df['CH_OPENING_PRICE'] = pd.to_numeric(stock_df['CH_OPENING_PRICE'], errors='coerce')
stock_df['CH_TRADE_HIGH_PRICE'] = pd.to_numeric(stock_df['CH_TRADE_HIGH_PRICE'], errors='coerce')
stock_df['CH_TRADE_LOW_PRICE'] = pd.to_numeric(stock_df['CH_TRADE_LOW_PRICE'], errors='coerce')

# Drop rows with NaN values in key columns after conversion
stock_df = stock_df.dropna(subset=['CH_CLOSING_PRICE', 'CH_OPENING_PRICE', 'CH_TRADE_HIGH_PRICE', 'CH_TRADE_LOW_PRICE'])

# Function to calculate RSI
def calculate_rsi(data, window=14):
    delta = pd.Series(data).diff(1)
    gain = (delta.where(delta > 0, 0)).rolling(window=window, min_periods=1).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window, min_periods=1).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Calculate RSI
stock_df['RSI'] = calculate_rsi(stock_df['CH_CLOSING_PRICE'])

# Function to identify divergences with loosened criteria
def identify_divergences(df, window=3):
    positive_divergences = []
    negative_divergences = []
    for i in range(window, len(df)):
        # Positive Divergence
        if df['CH_TRADE_LOW_PRICE'][i] < df['CH_TRADE_LOW_PRICE'][i - window] and df['RSI'][i] > df['RSI'][i - window]:
            positive_divergences.append((df['mTIMESTAMP'][i], df['CH_TRADE_LOW_PRICE'][i], df['RSI'][i]))
        # Negative Divergence
        if df['CH_TRADE_HIGH_PRICE'][i] > df['CH_TRADE_HIGH_PRICE'][i - window] and df['RSI'][i] < df['RSI'][i - window]:
            negative_divergences.append((df['mTIMESTAMP'][i], df['CH_TRADE_HIGH_PRICE'][i], df['RSI'][i]))
    return positive_divergences, negative_divergences

# Identify divergences with loosened criteria
positive_divergences, negative_divergences = identify_divergences(stock_df)

# Print detected divergences
print("Positive Divergences:", positive_divergences)
print("Negative Divergences:", negative_divergences)

# Prepare DataFrame to save to Excel
positive_divergence_data = {
    'Date': [date for date, price, rsi in positive_divergences],
    'Price Low': [price for date, price, rsi in positive_divergences],
    'RSI High': [rsi for date, price, rsi in positive_divergences]
}
negative_divergence_data = {
    'Date': [date for date, price, rsi in negative_divergences],
    'Price High': [price for date, price, rsi in negative_divergences],
    'RSI Low': [rsi for date, price, rsi in negative_divergences]
}
positive_divergence_df = pd.DataFrame(positive_divergence_data)
negative_divergence_df = pd.DataFrame(negative_divergence_data)

# Save to Excel file
with pd.ExcelWriter('SBIN_divergences.xlsx') as writer:
    positive_divergence_df.to_excel(writer, sheet_name='Positive Divergences', index=False)
    negative_divergence_df.to_excel(writer, sheet_name='Negative Divergences', index=False)

# Print confirmation
print("Positive and negative divergence data saved to SBIN_divergences.xlsx")

# Plot the closing price and RSI with divergences highlighted
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

# Plot Closing Price
ax1.plot(stock_df['mTIMESTAMP'], stock_df['CH_CLOSING_PRICE'], label='Closing Price', color='blue')
ax1.set_ylabel('Closing Price')
ax1.set_title(f'Divergences in {symbol}')
ax1.legend()
ax1.grid()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gcf().autofmt_xdate()

# Highlight positive divergences on price plot
ax1.scatter([date for date, price, rsi in positive_divergences],
            [price for date, price, rsi in positive_divergences],
            label='Positive Divergence', marker='^', color='green')

# Highlight negative divergences on price plot
ax1.scatter([date for date, price, rsi in negative_divergences],
            [price for date, price, rsi in negative_divergences],
            label='Negative Divergence', marker='v', color='red')

# Plot RSI
ax2.plot(stock_df['mTIMESTAMP'], stock_df['RSI'], label='RSI', color='purple')
ax2.axhline(70, linestyle='--', alpha=0.5, color='gray')
ax2.axhline(30, linestyle='--', alpha=0.5, color='gray')
ax2.set_ylabel('RSI')
ax2.set_xlabel('Date')
ax2.legend()
ax2.grid()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gcf().autofmt_xdate()

# Highlight positive divergences on RSI plot
ax2.scatter([date for date, price, rsi in positive_divergences],
            [rsi for date, price, rsi in positive_divergences],
            label='Positive Divergence', marker='^', color='green')

# Highlight negative divergences on RSI plot
ax2.scatter([date for date, price, rsi in negative_divergences],
            [rsi for date, price, rsi in negative_divergences],
            label='Negative Divergence', marker='v', color='red')

plt.tight_layout()
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), borderaxespad=0.)
plt.show()
```
