```python
# Calc Regular-ATR and SuperTrend-ATR
import pandas as pd
import nsepython as nse
import matplotlib.pyplot as plt


def calculate_atr(data, n):
  """Calculates the Average True Range (ATR) for a given window size."""
  if len(data) < n:
    raise ValueError("Data length must be greater than the window size.")

  atr = []
  for i in range(1, len(data)):
    tr = max(data['High'].iloc[i] - data['Low'].iloc[i],
             abs(data['High'].iloc[i] - data['Close'].iloc[i-1]),
             abs(data['Low'].iloc[i] - data['Close'].iloc[i-1]))
    if i < n:
      atr.append(tr)
    else:
      prev_atr = atr[i-1]
      atr.append((prev_atr * (n - 1) + tr) / n)

  return atr


def calculate_supertrend_atr(data, n):
  """Calculates the Supertrend ATR (modified ATR) for a given window size."""
  atr = calculate_atr(data, n)
  atr_df = pd.DataFrame(atr, columns=['ATR'])

  # Calculate basic upper and lower bands for Supertrend
  atr_df['Upper Band'] = data['High'] + (data['High'].rolling(window=n).max() - data['Low'].rolling(window=n).min()) * atr_df['ATR'] / 100
  atr_df['Lower Band'] = data['Low'] - (data['High'].rolling(window=n).max() - data['Low'].rolling(window=n).min()) * atr_df['ATR'] / 100

  return atr_df


# Define symbols and series
stock_symbol = "SBIN"
series = "EQ"
start_date = input("Enter start date (dd-mm-yyyy): ")
end_date = input("Enter end date (dd-mm-yyyy): ")

# Get stock data
try:
  data = nse.get_history(symbol=stock_symbol, series=series, start_date=start_date, end_date=end_date)
  df = pd.DataFrame(data)
except (nsepython.NsepyException, ConnectionError) as e:
  print(f"Error retrieving data: {e}")
  exit(1)

# Calculate Regular ATR (n=14)
try:
  regular_atr = calculate_atr(df, n=14)
except ValueError as e:
  print(f"Error calculating Regular ATR: {e}")
  exit(1)

# Save Regular ATR to Excel
regular_atr_df = pd.DataFrame(regular_atr, columns=['Regular ATR'])
regular_atr_df.to_excel(f"SBIN_Regular_ATR_{start_date}_{end_date}.xlsx", index=False)
print("Regular ATR saved to SBIN_Regular_ATR_{start_date}_{end_date}.xlsx")

# Calculate Supertrend ATR (n=10)
try:
  supertrend_atr = calculate_supertrend_atr(df.copy(), n=10)  # Copy data to avoid modifying original
except ValueError as e:
  print(f"Error calculating Supertrend ATR: {e}")
  exit(1)

# Save Supertrend ATR to Excel
supertrend_atr.to_excel(f"SBIN_Supertrend_ATR_{start_date}_{end_date}.xlsx", index=False)
print("Supertrend ATR saved to SBIN_Supertrend_ATR_{start_date}_{end_date}.xlsx")

# Plot Regular ATR
plt.figure(figsize=(12, 6))
plt.plot(df.index, regular_atr, label='Regular ATR')
plt.xlabel('Date')
plt.ylabel('ATR')
plt.title(f"SBIN Regular ATR (n=14) ({start_date} - {end_date})")
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()

# Plot Supertrend ATR (Upper and Lower Bands)
plt.figure(figsize=(12, 6))
```

```python
#Calc Supertrend
import pandas as pd
import nsepython as nse
import matplotlib.pyplot as plt
import numpy as np

# Define symbols and series
stock_symbol = input()
series = "EQ"
start_date = input()
end_date = input()

# Fetch equity history for SBIN
stock_data = nse.equity_history(stock_symbol, series, start_date, end_date)
stock_df = pd.DataFrame(stock_data)
stock_df['mTIMESTAMP'] = pd.to_datetime(stock_df['mTIMESTAMP'])
stock_df = stock_df.sort_values(by='mTIMESTAMP')

# Convert necessary columns to numeric
stock_df['CH_CLOSING_PRICE'] = pd.to_numeric(stock_df['CH_CLOSING_PRICE'], errors='coerce')
stock_df['CH_OPENING_PRICE'] = pd.to_numeric(stock_df['CH_OPENING_PRICE'], errors='coerce')
stock_df['CH_TRADE_HIGH_PRICE'] = pd.to_numeric(stock_df['CH_TRADE_HIGH_PRICE'], errors='coerce')
stock_df['CH_TRADE_LOW_PRICE'] = pd.to_numeric(stock_df['CH_TRADE_LOW_PRICE'], errors='coerce')

# Drop rows with NaN values in key columns after conversion
stock_df = stock_df.dropna(subset=['CH_CLOSING_PRICE', 'CH_OPENING_PRICE', 'CH_TRADE_HIGH_PRICE', 'CH_TRADE_LOW_PRICE'])

# Calculate True Range (TR)
stock_df['H-L'] = stock_df['CH_TRADE_HIGH_PRICE'] - stock_df['CH_TRADE_LOW_PRICE']
stock_df['H-CP'] = abs(stock_df['CH_TRADE_HIGH_PRICE'] - stock_df['CH_CLOSING_PRICE'].shift(1))
stock_df['L-CP'] = abs(stock_df['CH_TRADE_LOW_PRICE'] - stock_df['CH_CLOSING_PRICE'].shift(1))
stock_df['TR'] = stock_df[['H-L', 'H-CP', 'L-CP']].max(axis=1)

# Calculate Supertrend
n_supertrend = 10
multiplier = 10

stock_df['ATR'] = stock_df['TR'].rolling(window=n_supertrend).mean()

stock_df['Upper Basic'] = (stock_df['CH_TRADE_HIGH_PRICE'] + stock_df['CH_TRADE_LOW_PRICE']) / 2 + (multiplier * stock_df['ATR'])
stock_df['Lower Basic'] = (stock_df['CH_TRADE_HIGH_PRICE'] + stock_df['CH_TRADE_LOW_PRICE']) / 2 - (multiplier * stock_df['ATR'])

stock_df['Upper Band'] = stock_df['Upper Basic']
stock_df['Lower Band'] = stock_df['Lower Basic']

for i in range(1, len(stock_df)):
    if stock_df['CH_CLOSING_PRICE'][i-1] > stock_df['Upper Band'][i-1]:
        stock_df['Upper Band'][i] = max(stock_df['Upper Basic'][i], stock_df['Upper Band'][i-1])
    else:
        stock_df['Upper Band'][i] = stock_df['Upper Basic'][i]

    if stock_df['CH_CLOSING_PRICE'][i-1] < stock_df['Lower Band'][i-1]:
        stock_df['Lower Band'][i] = min(stock_df['Lower Basic'][i], stock_df['Lower Band'][i-1])
    else:
        stock_df['Lower Band'][i] = stock_df['Lower Basic'][i]

stock_df['Supertrend'] = np.nan
for i in range(1, len(stock_df)):
    if stock_df['CH_CLOSING_PRICE'][i-1] <= stock_df['Upper Band'][i-1]:
        stock_df['Supertrend'][i] = stock_df['Lower Band'][i]
    else:
        stock_df['Supertrend'][i] = stock_df['Upper Band'][i]



# Plot Supertrend
plt.figure(figsize=(14, 7))
plt.plot(stock_df['mTIMESTAMP'], stock_df['CH_CLOSING_PRICE'], label='Closing Price', color='blue')
plt.plot(stock_df['mTIMESTAMP'], stock_df['Supertrend'], label='Supertrend', color='red')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title(f'Supertrend of {stock_symbol}')
plt.legend()
plt.grid()

# Formatting date labels
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gcf().autofmt_xdate()

plt.show()


# Calculate Relative Vigor Index (RVGI)
stock_df['Close-Open'] = stock_df['CH_CLOSING_PRICE'] - stock_df['CH_OPENING_PRICE']
stock_df['Close-Open_1'] = stock_df['Close-Open'].shift(1)
stock_df['Close-Open_2'] = stock_df['Close-Open'].shift(2)
stock_df['Close-Open_3'] = stock_df['Close-Open'].shift(3)

stock_df['High-Low'] = stock_df['CH_TRADE_HIGH_PRICE'] - stock_df['CH_TRADE_LOW_PRICE']
stock_df['High-Low_1'] = stock_df['High-Low'].shift(1)
stock_df['High-Low_2'] = stock_df['High-Low'].shift(2)
stock_df['High-Low_3'] = stock_df['High-Low'].shift(3)

stock_df['RVGI'] = (stock_df['Close-Open'] + 2*stock_df['Close-Open_1'] + 2*stock_df['Close-Open_2'] + stock_df['Close-Open_3']) / (stock_df['High-Low'] + 2*stock_df['High-Low_1'] + 2*stock_df['High-Low_2'] + stock_df['High-Low_3'])
stock_df['RVGI_Signal'] = stock_df['RVGI'].rolling(window=4).mean()

# Plot RVGI
plt.figure(figsize=(14, 7))
plt.plot(stock_df['mTIMESTAMP'], stock_df['RVGI'], label='RVGI', color='green')
plt.plot(stock_df['mTIMESTAMP'], stock_df['RVGI_Signal'], label='RVGI Signal', color='orange')
plt.xlabel('Date')
plt.ylabel('RVGI Value')
plt.title(f'Relative Vigor Index (RVGI) of {stock_symbol}')
plt.legend()
plt.grid()

# Formatting date labels
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gcf().autofmt_xdate()

plt.show()

# Save Supertrend and RVGI values to Excel file
output_file = 'SBIN_Supertrend_RVGI.xlsx'
stock_df.to_excel(output_file, index=False, columns=['mTIMESTAMP', 'Supertrend', 'RVGI', 'RVGI_Signal'])
print(f"Supertrend and RVGI values have been saved to {output_file}")
```

```python
import pandas as pd
import nsepython as nse
import matplotlib.pyplot as plt
import openpyxl  # For Excel export

def calculate_supertrend(data, window=10, multiplier=3):
  """Calculates the Supertrend indicator for a given window and multiplier."""
  if len(data) < window:
    raise ValueError("Data length must be greater than the window size.")

  basic_ub = (data['CH_TRADE_HIGH_PRICE'] + data['CH_TRADE_LOW_PRICE']) / 2 + multiplier * calculate_atr(data, window)
  basic_lb = (data['CH_TRADE_HIGH_PRICE'] + data['CH_TRADE_LOW_PRICE']) / 2 - multiplier * calculate_atr(data, window)

  supertrend = []
  for i in range(len(data)):
    if i == 0:
      supertrend.append(None)
    elif data['CH_CLOSING_PRICE'].iloc[i-1] > basic_ub.iloc[i-1]:
      supertrend.append(basic_ub.iloc[i-1])
    elif data['CH_CLOSING_PRICE'].iloc[i-1] < basic_lb.iloc[i-1]:
      supertrend.append(basic_lb.iloc[i-1])
    else:
      supertrend.append(supertrend[i-1])

  # Final Supertrend calculation
  for i in range(len(data)):
    if supertrend[i] is not None:
      if data['CH_CLOSING_PRICE'].iloc[i] > supertrend[i]:
        supertrend[i] = basic_ub.iloc[i]
      elif data['CH_CLOSING_PRICE'].iloc[i] < supertrend[i]:
        supertrend[i] = basic_lb.iloc[i]

  return supertrend

def calculate_atr(data, window=14):
  """Calculates the Average True Range (ATR) for a given window."""
  if len(data) < window:
    raise ValueError("Data length must be greater than the window size.")

  atr = []
  for i in range(1, len(data)):
    tr = max(data['CH_TRADE_HIGH_PRICE'].iloc[i] - data['CH_TRADE_LOW_PRICE'].iloc[i],
             abs(data['CH_TRADE_HIGH_PRICE'].iloc[i] - data['CH_CLOSING_PRICE'].iloc[i-1]),
             abs(data['CH_TRADE_LOW_PRICE'].iloc[i] - data['CH_CLOSING_PRICE'].iloc[i-1]))
    if i < window:
      atr.append(tr)
    else:
      prev_atr = atr[i-1]
      atr.append((prev_atr * (window - 1) + tr) / window)

  return atr

# Define symbols and series
stock_symbol = "SBIN"
series = "EQ"
start_date = input()
end_date = input()

# Get stock data
data = equity_history(symbol=stock_symbol, series=series, start_date=start_date, end_date=end_date)
df = pd.DataFrame(data)


# Calculate Supertrend
supertrend = calculate_supertrend(df)

# Print Supertrend values
print("\nSupertrend Values:")
for i in range(len(supertrend)):
  print(f"{df.index[i]}: {supertrend[i]}")

# Optional: Plot Supertrend against dates (uncomment to visualize)
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['CH_CLOSING_PRICE'], label='Close')
plt.plot(df.index, supertrend, label='Supertrend')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title(f"SBIN Stock Price with Supertrend ({start_date} - {end_date})")
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()

# Optional: Save Supertrend to Excel (uncomment to export)
wb = openpyxl.Workbook()
ws = wb.active
ws.title = f"Supertrend_{stock_symbol}"

```
