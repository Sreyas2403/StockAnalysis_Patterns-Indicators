```python
from nsetools import Nse
import pandas as pd

start_time = input()
end_time = input()
symbol = input()
series = 'EQ'

def calculate_macd(data, short_window=12, long_window=26, signal_window=9):
    short_ema = data['CH_CLOSING_PRICE'].ewm(span=short_window, adjust=False).mean()
    long_ema = data['CH_CLOSING_PRICE'].ewm(span=long_window, adjust=False).mean()
    macd_line = short_ema - long_ema
    signal_line = macd_line.ewm(span=signal_window, adjust=False).mean()

    # Calculate MACD crossover points
    macd_cross_above_signal = (macd_line > signal_line) & (macd_line.shift(1) <= signal_line.shift(1))
    macd_cross_below_signal = (macd_line < signal_line) & (macd_line.shift(1) >= signal_line.shift(1))

    # Calculate 0-line crossover points
    macd_cross_above_zero = (macd_line > 0) & (macd_line.shift(1) <= 0)
    macd_cross_below_zero = (macd_line < 0) & (macd_line.shift(1) >= 0)

    return macd_line, signal_line, macd_cross_above_signal, macd_cross_below_signal, macd_cross_above_zero, macd_cross_below_zero

def calculate_rsi_macd(symbol,series, start_time, end_time):
    nse = Nse()
    equity_data = equity_history(symbol, series, start_date, end_date)

    # Convert data to DataFrame and sort by date
    data = pd.DataFrame(equity_data)
    data['mTIMESTAMP'] = pd.to_datetime(data['mTIMESTAMP'])
    data = data.sort_values(by='mTIMESTAMP')

    # Calculate MACD
    data['MACD'], data['Signal'], data['MACD_Cross_Above_Signal'], data['MACD_Cross_Below_Signal'], data['MACD_Cross_Above_Zero'], data['MACD_Cross_Below_Zero'] = calculate_macd(data)

    return data[['mTIMESTAMP', 'CH_CLOSING_PRICE', 'MACD', 'Signal', 'MACD_Cross_Above_Signal', 'MACD_Cross_Below_Signal', 'MACD_Cross_Above_Zero', 'MACD_Cross_Below_Zero']]

result = calculate_rsi_macd(symbol, series, start_time, end_time)
if result is not None:
    print(result)
```

```python
# Dates when MACD crossed Signal from below (bullish crossover)
#Dates when MACD crossed Signal from above (bearish crossover)
#Dates when MACD crossed 0-Line from below

from nsetools import Nse
import pandas as pd

start_time = input("Enter date: ")
end_time = input( "Enter date: ")
symbol = input("Enter symbol: ")

def calculate_macd(data, short_window=12, long_window=26, signal_window=9):
    short_ema = data['CH_CLOSING_PRICE'].ewm(span=short_window, adjust=False).mean()
    long_ema = data['CH_CLOSING_PRICE'].ewm(span=long_window, adjust=False).mean()
    macd_line = short_ema - long_ema
    signal_line = macd_line.ewm(span=signal_window, adjust=False).mean()
    return macd_line, signal_line

def calculate_macd_signals(symbol, start_time, end_time):
    nse = Nse()
    equity_data = equity_history(symbol, 'EQ', start_time, end_time)

    # Convert data to DataFrame and sort by date
    data = pd.DataFrame(equity_data)
    data['mTIMESTAMP'] = pd.to_datetime(data['mTIMESTAMP'])
    data = data.sort_values(by='mTIMESTAMP')

    # Calculate MACD and Signal
    data['MACD'], data['Signal'] = calculate_macd(data)

    # Calculate crossover points
    crossover_below_dates = []
    crossover_above_dates = []
    crossover_0line_below_dates = []

    prev_macd = data['MACD'].iloc[0]
    prev_signal = data['Signal'].iloc[0]

    for index, row in data.iterrows():
        if row['MACD'] > row['Signal'] and prev_macd <= prev_signal:
            crossover_below_dates.append(row['mTIMESTAMP'])
        elif row['MACD'] < row['Signal'] and prev_macd >= prev_signal:
            crossover_above_dates.append(row['mTIMESTAMP'])

        if row['MACD'] > 0 and prev_macd <= 0:
            crossover_0line_below_dates.append(row['mTIMESTAMP'])

        prev_macd = row['MACD']
        prev_signal = row['Signal']


    return crossover_below_dates, crossover_above_dates, crossover_0line_below_dates



crossover_below, crossover_above, crossover_0line_below = calculate_macd_signals(symbol, start_time, end_time)

print("Dates when MACD crossed Signal from below (bullish crossover):")
print(crossover_below)

print("Dates when MACD crossed Signal from above (bearish crossover):")
print(crossover_above)

print("Dates when MACD crossed 0-Line from below:")
print(crossover_0line_below)

```

```python
#Dates when RSI crossed RSI-based MA from below (bullish crossover)
#Dates when RSI crossed RSI-based MA from above (bearish crossover)
import nsepython
import pandas_ta as ta

symbol = input()
series = "EQ"
start_date = input()
end_date =  input()

# Get equity history
history = equity_history(symbol, series, start_date, end_date)

# Calculate RSI
history['RSI'] = ta.rsi(history['CH_CLOSING_PRICE'], length=14)

# Calculate RSI-based Moving Average (RSI MA)
rsi_ma_period = 9  # You can adjust this period as needed
history['RSI_MA'] = ta.sma(history['RSI'], length=rsi_ma_period)

# Identify crossover points
crossover_below_dates = []
crossover_above_dates = []

prev_rsi = None
prev_rsi_ma = None

for index, row in history.iterrows():
    if row['RSI'] is not None and row['RSI_MA'] is not None:
        if row['RSI'] > row['RSI_MA'] and prev_rsi is not None and prev_rsi <= prev_rsi_ma:
            crossover_below_dates.append(row['mTIMESTAMP'])
        elif row['RSI'] < row['RSI_MA'] and prev_rsi is not None and prev_rsi >= prev_rsi_ma:
            crossover_above_dates.append(row['mTIMESTAMP'])

        prev_rsi = row['RSI']
        prev_rsi_ma = row['RSI_MA']

print("Dates when RSI crossed RSI-based MA from below (bullish crossover):")
print(crossover_below_dates)

print("Dates when RSI crossed RSI-based MA from above (bearish crossover):")
print(crossover_above_dates)
```
