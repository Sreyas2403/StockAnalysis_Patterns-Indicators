```python
import pandas as pd
import nsepython as nse
import os

# taking inputs
symbol = input("Enter the stock symbol: ")
specific_date = input("Enter the specific date (dd-mm-yyyy): ")

def update_stock_data(symbol, specific_date):
    # Define series and file name
    series = "EQ"
    file_name = f"Price_{symbol}.xlsx"

    # Convert specific_date to the required format for NSE API
    specific_date_formatted = pd.to_datetime(specific_date, format='%d-%m-%Y').strftime('%d-%m-%Y')

    # Fetch the stock data for the specific date
    try:
        equity_data = nse.equity_history(symbol, series, specific_date_formatted, specific_date_formatted)
        new_data_df = pd.DataFrame(equity_data)

        # Convert necessary columns to appropriate formats
        new_data_df['mTIMESTAMP'] = pd.to_datetime(new_data_df['mTIMESTAMP'])
        new_data_df['CH_CLOSING_PRICE'] = pd.to_numeric(new_data_df['CH_CLOSING_PRICE'], errors='coerce')
        new_data_df['CH_TRADE_LOW_PRICE'] = pd.to_numeric(new_data_df['CH_TRADE_LOW_PRICE'], errors='coerce')
        new_data_df['CH_TRADE_HIGH_PRICE'] = pd.to_numeric(new_data_df['CH_TRADE_HIGH_PRICE'], errors='coerce')
        new_data_df['CH_OPENING_PRICE'] = pd.to_numeric(new_data_df['CH_OPENING_PRICE'], errors='coerce')
        new_data_df = new_data_df.dropna(subset=['CH_CLOSING_PRICE', 'CH_TRADE_LOW_PRICE', 'CH_TRADE_HIGH_PRICE', 'CH_OPENING_PRICE'])

        # Check if the file exists
        if os.path.exists(file_name):
            # Read the existing data from the Excel file
            existing_data_df = pd.read_excel(file_name)

            # Append the new data to the existing data, avoiding duplicates
            combined_data_df = pd.concat([existing_data_df, new_data_df]).drop_duplicates(subset=['mTIMESTAMP']).reset_index(drop=True)
        else:
            # If the file does not exist, use the new data as the combined data
            combined_data_df = new_data_df

        # Save the combined data to the Excel file
        combined_data_df.to_excel(file_name, index=False)
        print(f"Data for {specific_date} has been updated in {file_name}")

    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage

update_stock_data(symbol, specific_date)
```

```python
import pandas as pd
import nsepython

# code to fetch initial data by taking inputs
symbol = input()
series = 'EQ'
start_date = '01-01-2019'
end_date = '26-05-2024'
equity_data = equity_history(symbol, series, start_date, end_date)
stock_df = pd.DataFrame(equity_data)
excel_file = "Price_SBIN.xlsx"
stock_df.to_excel(excel_file, index=False)

# Input the specific date
specific_date_str = input("Enter the specific date (dd-mm-yyyy): ")
specific_date = pd.to_datetime(specific_date_str, format='%d-%m-%Y')

def append_stock_data(symbol, series, date, excel_file):
    # Convert the date to the required format
    start_date = end_date = date.strftime('%d-%m-%Y')

    # Fetch the equity history for the specific date
    equity_data = equity_history(symbol, series, start_date, end_date)

    # Convert to DataFrame
    new_data = pd.DataFrame(equity_data)

    # Load the existing Excel file
    existing_data = pd.read_excel(excel_file)

    # Append new data to the existing DataFrame
    updated_data = pd.concat([existing_data, new_data], ignore_index=True)
    print(updated_data)

    # Save the updated DataFrame back to Excel
    updated_data.to_excel(excel_file, index=False)


# Call the function to append data for the specific date
append_stock_data(symbol, series, specific_date, excel_file)
```
