```python
#Minimum and Maximum Volume

import pandas as pd

symbol = input( "Enter the symbol: ")
series = "EQ"
start_date = input("Enter the start date (dd-mm-yyyy): ")
end_date = input( "Enter the end date (dd-mm-yyyy): ")
equity_data = equity_history(symbol, series, start_date, end_date)

# Convert the equity data to a DataFrame
equity_df = pd.DataFrame(equity_data)

# Find minimum and maximum volumes along with their dates
min_volume_row = equity_df.loc[equity_df['CH_TOT_TRADED_QTY'].idxmin()]
max_volume_row = equity_df.loc[equity_df['CH_TOT_TRADED_QTY'].idxmax()]

min_volume = min_volume_row['CH_TOT_TRADED_QTY']
min_volume_date = min_volume_row['mTIMESTAMP']

max_volume = max_volume_row['CH_TOT_TRADED_QTY']
max_volume_date = max_volume_row['mTIMESTAMP']

print(f"Minimum Volume: {min_volume} on {min_volume_date}")
print(f"Maximum Volume: {max_volume} on {max_volume_date}")
```

```python
result_df = pd.DataFrame({
    'Volume': [min_volume, max_volume],
    'Date': [min_volume_date, max_volume_date],
    'Type': ['Minimum Volume', 'Maximum Volume']
})

# Save the results to an Excel file
excel_file = f'{symbol}_Volume_Info.xlsx'
result_df.to_excel(excel_file, index=False)

print(f'Results saved to {excel_file}')
```

```python
#Average Volume of 30 days(Rolling Average)
import pandas as pd

symbol = input( "Enter the symbol: ")
series = "EQ"
start_date = input( "Enter the start date (dd-mm-yyyy): ")
end_date = input( "Enter the end date (dd-mm-yyyy): ")
equity_data = equity_history(symbol, series, start_date, end_date)

# Convert the equity data to a DataFrame
equity_df = pd.DataFrame(equity_data)

# Convert the 'date' column to datetime format
equity_df['mTIMESTAMP'] = pd.to_datetime(equity_df['mTIMESTAMP'])

# Sort the data by date
equity_df = equity_df.sort_values(by='mTIMESTAMP')

# Calculate the rolling average volume over a 30-day window, handling outliers
equity_df['rolling_avg_volume'] = equity_df['CH_TOT_TRADED_QTY'].rolling(window=30, min_periods=1).mean()

# Print the average volume of 30 days
average_volume_30_days = equity_df['rolling_avg_volume'].mean()
print(f"Average Volume of 30 days: {average_volume_30_days}")
```

```python
result_df = pd.DataFrame({
    'Volume': [average_volume_30_days,],
    'Type': ['Avg Volume(30days)']
})

# Save the results to an Excel file
excel_file = f'{symbol}_Avg_Volume_.xlsx'
result_df.to_excel(excel_file, index=False)

print(f'Results saved to {excel_file}')
```

```python
#dates of exceeded volume than average volume
#corresponding closing prices of exceeded volume rows
#prior volume and prior closing price
#after volume and after cloing price

import pandas as pd

symbol = input( "Enter the symbol: ")
series = "EQ"
start_date = input( "Enter the start date (dd-mm-yyyy): ")
end_date = input( "Enter the end date (dd-mm-yyyy): ")
equity_data = equity_history(symbol, series, start_date, end_date)

# Convert the equity data to a DataFrame
equity_df = pd.DataFrame(equity_data)

# Convert the 'date' column to datetime format
equity_df['mTIMESTAMP'] = pd.to_datetime(equity_df['mTIMESTAMP'])

# Sort the data by date
equity_df = equity_df.sort_values(by='mTIMESTAMP')

# Calculate the rolling average volume over a one-year window
equity_df['rolling_avg_volume'] = equity_df['CH_TOT_TRADED_QTY'].rolling(window=252, min_periods=1).mean()

# Find the dates when the average volume of one year is exceeded
exceeded_dates = equity_df[equity_df['CH_TOT_TRADED_QTY'] > equity_df['rolling_avg_volume']]['mTIMESTAMP']

# Find the corresponding prices, volumes, and dates
exceeded_data = pd.DataFrame({
    'Exceeded Date': exceeded_dates,
    'Closing Price': equity_df.loc[equity_df['mTIMESTAMP'].isin(exceeded_dates), 'CH_CLOSING_PRICE'],
    'Prior Price': exceeded_dates.apply(lambda date: equity_df.loc[equity_df['mTIMESTAMP'] == date, 'CH_CLOSING_PRICE'].iloc[-1]),
    'Prior Volume': exceeded_dates.apply(lambda date: equity_df.loc[equity_df['mTIMESTAMP'] == date, 'CH_TOT_TRADED_QTY'].iloc[-1]),
    'After Price': exceeded_dates.apply(lambda date: equity_df.loc[equity_df['mTIMESTAMP'] == date, 'CH_CLOSING_PRICE'].iloc[0]),
    'After Volume': exceeded_dates.apply(lambda date: equity_df.loc[equity_df['mTIMESTAMP'] == date, 'CH_TOT_TRADED_QTY'].iloc[0]),
})

exceeded_data.to_excel('exceeded_data.xlsx', index=False)
print(exceeded_data)

print("exceeded_data saved to exceeded_data.xlsx" )
```

```python
import nsepython
import pandas_ta as ta

symbol = "SBIN"
series = "EQ"
start_date = "01-01-2023"
end_date = "31-12-2023"

# Get equity history
history = equity_history(symbol=symbol, series=series, start_date=start_date, end_date=end_date)

# Calculate RSI
history['RSI'] = ta.rsi(history['CH_CLOSING_PRICE'], length=14)

# Calculate RSI-based Moving Average (RSI MA)
rsi_ma_period = 9  #  adjust this period as needed
history['RSI_MA'] = ta.sma(history['RSI'], length=rsi_ma_period)

# Identify dates of RSI crossovers above RSI-MA within RSI range (20 <= RSI <= 40)
rsi_crossovers = []
prev_rsi_above_ma = False  # Initialize to False

for index, row in history.iterrows():
  rsi = row['RSI']
  rsi_ma = row['RSI_MA']

  # Check for RSI crossing above RSI-MA within RSI range (bullish crossover)
  if rsi > rsi_ma and 20 <= rsi <= 40 and (prev_rsi_above_ma is None or not prev_rsi_above_ma):
    rsi_crossovers.append(row['mTIMESTAMP'])
    prev_rsi_above_ma = True
  else:
    prev_rsi_above_ma = False  # Reset if not a crossover

# Print results
print(history[['mTIMESTAMP', 'CH_CLOSING_PRICE', 'RSI', 'RSI_MA']])
print("\nList of Dates when RSI crossed above RSI-MA (20 <= RSI <= 40):", rsi_crossovers)
```
